We are going to split our data into 2 categories, training data, test data

by using training data model(connections is made b/w feature and label) is created, by using test data model is tested for accuracy
 
 We are going to create 4 arrays from these 2 feature and label arrays using the library sklearn.model_selection method train_test_split

```python
from sklearn import datasets
from sklearn.model_selection import train_test_split

iris = datasets.load_iris()

features = iris.data
labels = iris.target

#spliting and giving half of the features to features_train and features test and same to labels
features_train, features_test, labels_train, labels_test = train_test_split(features, labels, test_size = 0.5)
```

```python
# sklearn.neighbors module has the bunch of classification algorithms in it
# we are going to use Kneighborclassifier for classification 
from sklearn.neighbors import KNeighborsClassifier
```

```python
# KneightborsClassifier will classify based the taking some neighbors using euclidean dist and based on the majority of the neighbor's class, it will clasify
my_classifier = KNeighborsClassifier() # create classifier and pass features_train and label train
```

```python
#take the classifier and using the method fit where we pass the training data to laad the training data to the model
my_classifier.fit(features_train, labels_train)
```

```python
# we'll predict the model
prediction = my_classifier.predict(features_test)
```

```python
prediction
```

```python
#we are going to find the accuracy of the model
from sklearn.metrics import accuracy_score # in this method we can pass two array(true_label, predicted_label) and it will give the accuracy of similaty b/w these two arrays
```

```python
accuracy_score(labels_test, prediction)
```

### KnearestNeighbor classfier from scratch

based on only one nearest neighbor we can predict the output, wheras in the previous it took 5 neighbors to predict

```python
from scipy.spatial import distance

def euc(a,b):
    return distance.euclidean(a,b)

class ScrappyKNN(): # 3functions, fit, predict, closest(to be used for computation only inside the class)
    def fit(self, features_train, labels_train):
        self.features_train = features_train
        self.labels_train = labels_train
        
    def predict(self, features_test):
        predictions = []
        for item in features_test:
            # determine which other point is closest and that is set to label value
            label = self.closest(item)
            predictions.append(label)
            
        return predictions
    
    def closest(self, item):
        best_dist = euc(item, self.features_train[0])
        best_index = 0
        for i in range(1, len(self.features_train)):
            dist = euc(item, self.features_train[i])
            if dist < best_dist:
                best_dist = dist
                best_index = i
        
        return self.labels_train[best_index]
```

```python
my_classifier = ScrappyKNN()
my_classifier.fit(features_train, labels_train)
```

```python
prediction = my_classifier.predict(features_test)
accuracy_score(labels_test, prediction)
```
