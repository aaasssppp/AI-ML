We are doing this method to avoid overfitting the data.

In this we'll randomly turn off (or) remove the neuron in the hidden layer at each iteration based on the given value like 0.5 -> 50% of neurons are turned off, etc.

Why will dropout help with overfitting?
1. It can't rely on one input as it might be dropped out at random
2. Neurons will not learn redundant details of inputs

In this, we are going to find whether it is rock or metal based on the sonar data

# Importing Library

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Reading Data

```python
df = pd.read_csv("sonar_dataset.csv", header=None)
df.sample(5)
```

### Shape of data

```python
df.shape
```

### Check whether the data contains any NaN values

```python
df.isna().sum() # since there is no NaN values, there is no need for data cleaning
```

### Total Columns

```python
df.columns # 60th is the output columns
```

### Different values in the output column

```python
df[60].value_counts()
```

### Inputs and Outputs

```python
x = df.drop(60,axis='columns')
y = df[60]
y.head() # it is in string, we have to convert it into numeric by one hot encoding
```

# One hot encoding

```python
y = pd.get_dummies(y, drop_first = True) # since there are only two cols and after removing one, it knows the values are binary, so even mentionaning dtype = int, it will default to boolean only. To overcome this we have to chage datatype as int after it gives us boolean
y = y.astype(int)
y
```

```python
y.value_counts()
```

# Data Spliting

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=1)
```

```python
x_train.shape
```

```python
x_test.shape
```

# Model Creation & Training

```python
import tensorflow as tf
from tensorflow import keras
```

```python
model = keras.Sequential([
    keras.layers.Dense(60, input_dim=60, activation='relu'), # hidden layer
    keras.layers.Dense(30, activation='relu'), # hidden layer
    keras.layers.Dense(15, activation='relu'), # hidden layer
    keras.layers.Dense(1, activation='sigmoid') # output layer
])

model.compile(
    loss = 'binary_crossentropy',
    optimizer='adam',
    metrics=['accuracy']
)

model.fit(x_train,y_train, batch_size=8, epochs=100) # mini-batch gradient descent. Each time the 8 batches runs and error and backpropogation occurs
```

Accuracy = 100 -> overfitting

# Model Evaluation

```python
model.evaluate(x_test,y_test)
```

Test accuracy -> 71% (due to overfitting)

# Model Prediction

```python
y_pred = model.predict(x_test).reshape(-1)
print(y_pred[:10])

# round the values to nearest integer i.e 0 0r 1
y_pred = np.round(y_pred)
print(y_pred[:10])
```

```python
y_test[:10]
```

# Classification report

```python
from sklearn.metrics import classification_report, confusion_matrix
print(classification_report(y_test,y_pred))
```

# Adding Droupout layer

Dropout layer can be added after every hidden layer

```python
modeld = keras.Sequential([
    keras.layers.Dense(60, input_dim=60, activation='relu'), # hidden layer
    keras.layers.Dropout(0.5), # Dropout layer - 50%
    keras.layers.Dense(30, activation='relu'), # hidden layer
    keras.layers.Dropout(0.5), # Dropout layer - 50%
    keras.layers.Dense(15, activation='relu'), # hidden layer
    keras.layers.Dropout(0.5), # Dropout layer - 50%
    keras.layers.Dense(1, activation='sigmoid') # output layer
])

modeld.compile(
    loss = 'binary_crossentropy',
    optimizer='adam',
    metrics=['accuracy']
)

modeld.fit(x_train,y_train, batch_size=8, epochs=100) # mini-batch gradient descent. Each time the 8 batches runs and error and backpropogation occurs
```

Now Training accuracy reduced means No overfitting

# Model Evaluation after Dropout layer

```python
modeld.evaluate(x_test, y_test)
```

Test accuracy increased

# Model Prediction

```python
y_pred = model.predict(x_test).reshape(-1)

# round the values to nearest integer i.e 0 0r 1
y_pred = np.round(y_pred)
print(y_pred[:10])
```

# Classification report after Dropout layer

```python
from sklearn.metrics import classification_report, confusion_matrix
print(classification_report(y_test,y_pred))
```
