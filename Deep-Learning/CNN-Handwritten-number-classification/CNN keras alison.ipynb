tensonflow is a popular ML library created by google, used by phd, data scientists. Because of that it can be very complicated for the beginners

keras is a high level neural network API, written in python running on top of tensorflow. Userfriendly, modularity(customisable), extendable(new modules are simple to add), works with python

Core data sturcture of keras is model. data -> model -> prediction

A type of model is sequential. It is a linear stack of layers

Convolutional Neural Networks(CNN or ConvNet)

These are easiest and best way to classify images. We can also train a model to play a game. For example a model is made to see YouTube cooking videos, it not only can identify knife, vegetables but also it learned how to use those

A toy ConvNet: X's and O's

Says whether it is an X or an O

CNN breaks the picture into pixel and evaluate

Tricker cases are translation(x or o not at the center but at the corner or anywhere), scalling(different size), rotation(rotated version), weight(bolder or thiner). It should find in all these cases

We can find the difference easily but the computers sees only the values of the pixels i.e 1, -1 

Take little features of the image

Feature match(Do this for every feature)

Filtering: 
1. line up the feature and the image patch
2. Mulitply each image pixel by the corresponding feaure pixel 
3. Add them up
4. Divide by the total number of pixels in the feature

Convolution: Trying every possible match

Convolutional layer: One image becomes a stack of filtered images

Pooling: Shrinking the image stack
1. Pick a window size (usually 2 or 3). It should be smaller than the filtering step
2. pick a stride(usually 2). Jump
3. Walk your window across your filtered images
4. From each window, take the maximum value.

Now the stack of images, now became stack of even smaller images

Normalization: Keet the math from breaking by tweaking each of the values just a bit. Change every negative into zero using Rectified Linear Units(ReLUs)

Now the stack of images becomes a stack of images with no negative values

Layers get stacked: The output of one becomes the input of the next.

Input -> 1. Convolution, 2. ReLU, 3. Pooling -> Stack of smaller images

Deep Stacking: Layers can be repeated several times

Input -> 1. Conv, 2. ReLU, 3. Conv, 4. ReLU, 5. Pooling, 6. Conv, 7. ReLU, 8. Pooling -> smaller image/data

Fully connected layer(keras - dense layer): Every value gets a vote:
Every data is squeced into a linear array. Vote depends on how strongly a value predicts X or O

based on the percentage of prediction of x and o, the model predicts whether it is x or o Eg. X=91% and O=51%, the model predicts it to X

For more precision, reduce the outputs of fully connected layer gradually by adding many fully connected layer

For example, to predict a number between 0-9, first Fully connected layer gives 30outputs and the second give 10 outputs and from that the model predicts.

Finally for predicting X or O

Example,

Input -> 1. Conv, 2. ReLU, 3. Conv, 4. ReLU, 5. Pooling, 6. Conv, 7. ReLU, 8. Pooling, 9. Fully connnected, 10. Fully connected -> X(92%) or O(51%)



### Preparing data

```python
import keras
from keras.datasets import mnist
```

```python
#data = mnist.load_data()
```

mnist - database of handwritten digits, 60k training images, 10k testing images

```python
#you don't have to remember this, you can always check the documentation
(features_train, labels_train), (features_test, labels_test) = mnist.load_data() 
#returns two tuples 
#x_train, x_test: unit8 array of shape [num_samples, 28(px), 28(px)], it is a grey scale, so we have to use the sa depth as 1
#y_train, y_test: uint8 arra of digit labels(int in range 0-9) with shape (num_samples)
```

```python
features_train.shape
```

6th March 2025

```python
#now we are going to reshape our train and test by adding depth 1
pixel_width = 28
pixel_height = 28
pixel_depth = 1
features_train = features_train.reshape(features_train.shape[0], 28, 28, 1)
features_test = features_test.reshape(features_test.shape[0], 28, 28, 1)
```

```python
#features_train.shape
```

```python
input_shape = (pixel_width, pixel_height, 1) #we want the input to be in this shape because it the requirement of the CNN
```

```python
#another requirement for CNN is the type of all variables should be in type 'float 32', So now we'll typecast
features_train = features_train.astype('float32')
features_test = features_test.astype('float32')
```

```python
#features_train
```

```python
#in the color representation, every pixel ranges from(0-255, 0-255, 0-255). We are going to convert it into percentage by dividing every value by 255
features_train /= 255
features_test /= 255
```

```python
print(labels_train[5])
```

```python
#changing to binary matrix
#our model will not tell the value of the labels_train = 2, whereas, it can classify into 2, like in a binary matrix(0/1), it should change the value of 3rd one to ON and rest to OFF
num_of_classes = 10
labels_train = keras.utils.to_categorical(labels_train, num_of_classes)
labels_test = keras.utils.to_categorical(labels_test, num_of_classes)
```

```python
labels_train[5]
```

### Using Sequential

```python
from keras.models import Sequential #sequential is one of 2 types of models in keras, 1. model, 2. sequential. In sequential you can add as many layers as you want and make it complex
from keras.layers import Conv2D, MaxPooling2D, Dense, Dropout, Flatten # it has convolutional layer, pooling layer, fully connected layer
```

```python
model = Sequential()
#using 3x3 window size to filter all over the image
#activate each value to be positive or zero
#input shape, 28x28x1
model.add(Conv2D(32, kernel_size=(3,3), activation = 'relu', input_shape=input_shape)) #taking 32 unique features to compare with and create 32 convoluted images. In the example we took 3features to find X or O
print("POST CONV2D: ", model.output_shape)
model.add(MaxPooling2D(pool_size=(2,2))) #pool size should be less than window size, we take it as 2 and stride size the same, so we don't have to mention
print("POST MaxPooling2D: ", model.output_shape)
model.add(Dropout(0.25))
print("POST Dropout: ", model.output_shape)
```

we want to model to find the trend -> Just right

when the model didn't see wide enough data, it will tell like it is 3 or not 3, not exact answer -> underfitting

when the model is trained with a lot of '3' samples, it will consider those 3's only to be 3, but if we feed any other 3, it won't recognize it -> overfitting

dropout layer is used to avoid overfitting, it set the random fraction of input to zero, so that it won't think like the answer should be the same as the training

```python
model.add(Flatten()) # making it into a single stack
print("POST Flatten: ", model.output_shape)
```

```python
model.add(Dense(128, activation = 'relu'))
print("POST Dense1: ", model.output_shape)
```

```python
model.add(Dense(num_of_classes, activation = 'softmax'))
print("POST Dense2: ", model.output_shape)                
```

Compiling the model

```python
model.compile(loss=keras.losses.categorical_crossentropy, optimizer=keras.optimizers.Adadelta(), metrics=['accuracy']) 

batch_size = 32 #set it equal to number of convoluted layers
epochs = 10 #10 cycles to make the model better
#verbose(validate/prints out) is set to 1, so that we can see it, which means binary ON
model.fit(features_train, labels_train, batch_size=batch_size, epochs=epochs, verbose=1, validation_data=(features_test, labels_test))
score = model.evaluate(features_test, labels_test, verbose=0)
```
