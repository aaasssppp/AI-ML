Few terms in SVM

Gamma: 
1. High Gamma, high regularization(C): High accuracy, complex separating plane, complex computation, overfitting
2. Low Gamma, low regularization(C): Less accuracy, less complex separating plane(lesser order equations), simple computation

For complex boundaries, we may go to higher dimension and will try to create a boundary(like circle etc. z= x^2 + y^2)

The transformation is called kernel

```python
import pandas as pd
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
dir(iris)
```

# Formating the data in dataframe

```python
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df
```

```python
df['target'] = iris.target
df
```

To get the flower name as a column we have to do a computation on the target column

# Compution on a col to get other column

It can be done using apply method

```python
df['flower_name'] = df.target.apply(lambda x: iris.target_names[x])
df.head(-10)
```

Create 3 dataframes for 3 flowers

```python
df0 = df[df['target'] == 0] # setosa
df1 = df[df['target'] == 1] # versicolor
df2 = df[df['target'] == 2] # virginica
```

# Visualization

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.xlabel('sepal length (cm)')
plt.xlabel('sepal width (cm)')
plt.scatter(df0['sepal length (cm)'], df0['sepal width (cm)'], color='green', marker='+')
plt.scatter(df1['sepal length (cm)'], df1['sepal width (cm)'], color='blue', marker='.')
plt.show()
```

```python
plt.xlabel('petal length (cm)')
plt.xlabel('petal width (cm)')
plt.scatter(df0['petal length (cm)'], df0['petal width (cm)'], color='green', marker='+')
plt.scatter(df1['petal length (cm)'], df1['petal width (cm)'], color='blue', marker='.')
plt.show()
```

# Spliting Data

```python
from sklearn.model_selection import train_test_split
inputs = df[['sepal length (cm)','sepal width (cm)','petal length (cm)','petal width (cm)']]
target = df.target
x_train, x_test, y_train, y_test = train_test_split(inputs,target,test_size=0.2)
```

# Model Creation

```python
from sklearn.svm import SVC #suppor vector classifier
svc = SVC()
```

# Model Training

```python
svc.fit(x_train, y_train)
```

# Model Accuracy

```python
svc.score(x_test,y_test) # above 90%
```
