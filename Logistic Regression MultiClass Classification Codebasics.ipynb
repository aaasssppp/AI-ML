```python
import matplotlib.pyplot as plt

from sklearn.datasets import load_digits # dataset containing 8x8 image of numbers, 1797 images
```

```python
digits = load_digits()
```

```python
dir(digits)
```

```python
digits.data[0]
```

```python
plt.gray() # Sets the colormap to grayscale, ensuring the image is displayed in black and white rather than a default color map.
plt.matshow(digits.images[0]) # plt.matshow() is used to visualize matrix-like data as an image.
# Displays the first digit image (digits.images[0]) from the dataset.
```

```python
digits.target[0:5] # label
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(digits.data, digits.target, test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
LR = LogisticRegression()
```

```python
LR.fit(x_train, y_train)
```

```python
LR.predict(x_test)
```

```python
LR.score(x_test,y_test) # 96.67%
```

# To see where our model didn't perform well

```python
y_predicted = LR.predict(x_test) # storing our prediction
from sklearn.metrics import confusion_matrix # we need confusion matrix which tells where our did mistake

cm = confusion_matrix(y_test,y_predicted)
cm # this is the confusion matrix
```

# Visualization of Confusion Matrix

```python
import seaborn as sns
plt.figure(figsize=(10,7))
sns.heatmap(cm, annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

0-0 = 29 -> 29 times the model predicted correctly, same goes to other

1-3 = 1 -> 1 time the model predicted 1 incorrectly as 3, same goes to other
