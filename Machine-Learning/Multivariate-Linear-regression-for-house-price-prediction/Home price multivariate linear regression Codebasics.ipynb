# Formula

price = m1 * area + m2 * bedrooms + m3 * age + b

# Importing libraries

```python
import numpy as np
import pandas as pd
from sklearn import linear_model
```

```python
# Data
```

```python
home_price = {
    'area': [2600,3000,3200,3600,4000],
    'bedrooms': [3,4,None,3,5],
    'age': [20,15,18,30,8],
    'price': [550000,565000,610000,595000,760000]
}
```

```python
df = pd.DataFrame(home_price, index=[x for x in range(len(home_price['area']))])
df
```

# Data cleaning

replace the NaN value by the median

```python
median_bedroom = int(df.bedrooms.median())
median_bedroom
```

```python
df.bedrooms = df.bedrooms.fillna(median_bedroom)
```

```python
df
```

# Model creation

```python
model = linear_model.LinearRegression()
model.fit(df[['area','bedrooms','age']],df.price)
```

# Model Parameters

```python
model.coef_ # m1,m2,m3
```

```python
model.intercept_ # b
```

# Model Prediction

```python
model.predict([[3000,3,40]]) # 3000sq ft, 3bedrooms, 40years old
```

```python
model.predict([[2500,3,5]])# 2500sq ft, 3bedrooms, 5year old
```
