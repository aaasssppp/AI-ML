In this we are going to find the m(slope) and b(y-intercept) using gradient descent

# Importing Library

```python
import numpy as np
import matplotlib.pyplot as plt
```

# Steps to take the iterations and learning_rate.
1. less iteration, change learning_rate and reduce cost
2. fix iteration, increase iteration

iteration = 10, learning rate = 0.001 -> works (cost is reducing)
iteration = 10, learning rate = 0.01 -> works
iteration = 10, learning rate = 0.1 -> not working (cost is increasing)
so learning rate should be between 0.1 and 0.001
0.09 is not working, but 0.08 is working
so we'll fix 0.08 and we'll increase the iteration to 10000 -> this gives us most accurate values of m,b
When we reach the global minima, our cost will became saturated(remain the same) after few iterations. As it is in the given example. So we can use floating point comparison, and we can stop the iteration once it got saturated

# Implementation and Visualization

```python
def gradient_descent(x,y):
    m_curr = b_curr = 0 # start with any value of m and b. Then take some baby step to find the correct value of m and b
    iterations = 10000 # number of steps we are going to take, we will fist take 1000 and the min cost we got is 0.478. so we'll increase the iterations to 10000 and our cost is very much reduced by 0.001
    n = len(x) # length of x and y
    learning_rate = 0.08 # start with a small value and change accordingly
    
    for i in range(iterations): # for these number of steps
        y_predicted = m_curr*x + b_curr # predict the value of y

        plt.plot(x,y_predicted,color='green') # ploting the lines in the graph
        
        cost = (1/n)*sum([val**2 for val in (y-y_predicted)]) # cost fn, cost should be reduced each step
        md = -(2/n)*sum(x*(y-y_predicted)) # m derivative, mse(cost_fn) = (1/n) sum((y-y_pred)^2) -> finding derivative of cost wrt m will give the metioned equation
        bd = -(2/n)*sum(y-y_predicted) # b derivative
        m_curr = m_curr - learning_rate*md # update current m value at each iteration
        b_curr = b_curr - learning_rate*bd # update current b value at each iteration
        print("m {}, b {}, cost {}, iteration {}".format(m_curr,b_curr,cost,i))
```

```python
x = np.array([1,2,3,4,5]) # y = m*x + b
y = np.array([5,7,9,11,13]) # y = 2x + 3; m=2, b=3
gradient_descent(x,y)
```
